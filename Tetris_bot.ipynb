```python
import os
from rl_zoo3.exp_manager import ExperimentManager
from stable_baselines3 import DQN

args = None
algo = "dqn"
env_id = "ALE/Tetris-v5"
log_folder = "logs/"

experiment_manager = ExperimentManager(args, algo, env_id, log_folder)

model_path = os.path.join('/content/logs/dqn/ALE-Tetris-v5_1/best_model.zip')

# Load the trained model
model = DQN.load(model_path)

experiment_manager.save_trained_model(model)
```
